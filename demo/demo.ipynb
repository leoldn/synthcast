```python
agents_per_country = 20
total_countries = 5

macro_questions = [
    "Do you think its a good idea to move to renewable energy and build more windfarms in your country?",
    "Do you think your government will impose tariffs on import from other countries?",
    "Do you think your government will require all new vehicles to be electric by 2035?",
    "Do you think your government will prioritise WTO agreements over local/regional billateral ageements?",
]

social_questions = [
    "Do you think your country will face significant domestic protests against government policies by 2030?",
    "Do you think your government will impose more taxes on citinzens in next five years",
    "Do you expect increase in crime rate in your country in next five years?",
    "Do you think high electricity pries justify net zero goals of your government?",
]

marketing_questions = [
    "Would you buy electric car if it was more expencive than the petrol vehicle?",
    "Would you want to pay for your children education if it is better quality?",
    "Would you pay for online services and media if they would not share your data with third parties?",
]
```

```python
generator = PersonaGenerator()

codes = [d["iso_code"] for d in generator.population_data.values()][:total_countries]
sim = Simulation({c: agents_per_country for c in codes}, temperature=0.7)
```

```python
for question in tqdm(macro_questions):
    res = sim.ask_question(question, save_results=True)

for question in tqdm(social_questions):
    res = sim.ask_question(question, save_results=True)

for question in tqdm(marketing_questions):
    res = sim.ask_question(question, save_results=True)
```
